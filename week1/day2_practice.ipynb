```python
import os
from dotenv import load_dotenv
```

```python
load_dotenv(override=True)
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

if OPENAI_API_KEY:
    print(f"Open AI key loaded successfully and starts with {OPENAI_API_KEY[:8]}")
else:
    print("Failed to load Open AI key. Please check your .env file.")
```

```python
import requests

base_url = "https://api.openai.com/v1/chat/completions"

headers = {"Authorization": f"Bearer {OPENAI_API_KEY}", "content-type": "application/json"}

messages = "Tell me a fun fact"

payload = {"model": "gpt-5-nano", "messages": [{"role": "user", "content": messages}]}

payload
```

```python
response = requests.post(base_url, headers=headers, json=payload)

response.json()
```

```python
response.json()["choices"][0]["message"]["content"]
```

```python
from openai import OpenAI

openai = OpenAI()
user_prompt = "Tell me a fun fact"
messages = [{"role": "user", "content": user_prompt}]

response = openai.chat.completions.create(model="gpt-5-nano", messages = messages)
response.choices[0].message.content
```

```python
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")

if GOOGLE_API_KEY:
    print(f"Google API key loaded successfully and starts with {GOOGLE_API_KEY[:3]}")
else:
    print("Google API key loading failed. Check .env file")
```

```python
load_dotenv(override=True)

GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"

gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=GOOGLE_API_KEY)

model = "gemini-2.5-flash"
user_prompt = "Tell me a fun fact"
messages = [{"role": "user", "content": user_prompt}]

response =gemini.chat.completions.create(model=model, messages=messages)
response.choices[0].message.content
```
