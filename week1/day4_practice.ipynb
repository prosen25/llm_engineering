```python
import tiktoken

encoding = tiktoken.encoding_for_model("gpt-4.1-mini")

tokens = encoding.encode("Hi, my name is Prosen.")

tokens
```

```python
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id} = {token_text}")
```

```python
import os
from dotenv import load_dotenv

load_dotenv(override=True)

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

if OPENAI_API_KEY:
    print(f"Openai API key loaded successfully and starts with {OPENAI_API_KEY[:8]}")
else:
    print("Cannot load Openai API key. Please check .env file!")
```

```python
from openai import OpenAI

openai = OpenAI()

messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "Hi, my name is Prosen."}
]

response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)

response.choices[0].message.content
```
